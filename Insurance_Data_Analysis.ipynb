##  **Numerical v/s Numerical**

```python
# (1) relation b/w "age" vs "bmi"

sns.jointplot(x = "age",y = "bmi",data = df,kind = "hex")
plt.show()
             
```

From above jointplot shows the normal BMI criteria is ~22 to ~48. All the age group of people are most in these range.Hence,most common age group are <25 maximum number of people lies in this range with normal BMI.

```python
# (2) finding and plotting correlation between all the numerical columns 

corr_mat = df.corr(numeric_only = True)
plt.figure(figsize = (10,5))
sns.heatmap(corr_mat, linewidths = 0.5,annot = True,fmt = ".1f")
plt.title("correlation for numerical data")
plt.show()
```

As per the correlation heatmap, Whose score < 0.2 are considered as weak relation and rest of all are quite good relation compare to them.

```python
# (3) finding relation between all two numerical columns with each other separatly 

numeric_data = ["age","bmi","children","charges"]
sns.pairplot(df[numeric_data]) 
plt.show()
```

Here,in above pairplot shows each numerical relation between all the numerical columns.
1) partial positive relation between columns : age vs charges
2) no relation columns : remain almost all the columns are not correlated/no relation between them 
* So, Only one relation that are said to be partially positive and rest of all had no relation or weak relation. 

#  **Categorical v/s Numerical**

```python
# (1) finding relation in between of region and their charges how they differ?

plt.figure(figsize = (10,8))
sns.boxplot(x = "region",y = "charges", data = df, hue = "region")
plt.title("Region vs Charges")
plt.show()
```

In this box plot,The median price of insurance in all the region is almost equal. Here, in southeast region some of people paying higher insurance compare to all of them and maximum number of outlier is also from southeast and southwest.

#  **Categorical v/s Categorical**

```python
# (1) finding relation between region and gender

cp = pd.crosstab(index = df["region"] , columns = df["sex"])
cp
```

```python
cp.plot(kind = "bar",stacked = True,figsize = (8,6),cmap  = "Paired")
plt.title("Region vs Sex")
plt.legend(loc = "upper left")
plt.show()
```

From above stacked bar plot it shows the relation of region and gender, For the population of female and male is almost equal in all the region excluding southeast region. In southeast region still female population is similar as all the region, but male population is higher in this region. 

# Check If The Number Of Premium Charges For Smokers Or Non-Smokers Is Increasing As They Are Aging?

```python
# Checking the premium charges for smoker or non-smoker is increasing as they are aging
```

```python
# figuring out the trend line for premium charges as they are aging (Generel Scenario)

plt.figure(figsize = (13,8))
sns.lineplot(x = "age", y = "charges", data = df, color = "green")
plt.title("Trend Line For The Age and Their Premium")
plt.show()
```

Line plot shows the positive trend line for age and theire premium charges are getting increase as they are aging.

```python
# Now checking premium for smoker or non-smoker by their age

# How charges are differ from smoker or non-smoker as their age is getting increase ?
```

```python
plt.figure(figsize = (8,7))
sns.scatterplot(x = "age",y = "charges",data = df,hue = "smoker")
plt.title("Premium For Smoker or Non-Smoker As They Are Aging")
plt.show()
```

From above scatter plot, It clearly tell us that the premium charges for the smoker is getting increases as they are aging and even higher premium charges with compare to non-smoker.
